```python
# Import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Read dataset into a DataFrame
dataset = pd.read_csv('/content/PlayTennis.csv')
```

```python
# Print the dataset
print("Dataset:")
print(dataset)
```

```python
dataset.head()
```

```python
# Define features and target variable
features = ['Outlook', 'Temperature', 'Humidity', 'Wind']
target = 'Play Tennis'

```

```python
# Convert categorical features into numerical values
dataset_encoded = pd.get_dummies(dataset[features])
```

```python
# Build decision tree
model = DecisionTreeClassifier(criterion='entropy')
model.fit(dataset_encoded, dataset[target])

```

```python
# Visualize decision tree
plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=dataset_encoded.columns, class_names=model.classes_, filled=True)
plt.show()

```

```python
# Classify a new sample
new_sample = pd.DataFrame({
    'Outlook': ['Sunny'],
    'Temperature': ['cCool'],
    'Humidity': ['Normal'],
    'Wind': ['Weak']
})
new_sample_encoded = pd.get_dummies(new_sample)

# Get feature names used during training
feature_names_training = dataset_encoded.columns

# Align feature names of new sample with feature names used during training
new_sample_encoded_aligned = new_sample_encoded.reindex(columns=feature_names_training, fill_value=0)

# Classify the aligned new sample
classification = model.predict(new_sample_encoded_aligned)
print("Classification:", classification)
```
